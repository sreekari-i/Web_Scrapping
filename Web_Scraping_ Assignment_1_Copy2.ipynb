## 1. Python program to display all the header tags from "en.wikipedia.org/wiki/Main_Page".

```python
import pandas as pd
All_headers = pd.DataFrame({})
All_headers['Header Name']=all_headers
All_headers
```

# 2. Python program to display IMDB’s Top rated 100 movies’ data (i.e. Name,IMDB rating, Year of release) and saving it in form of a CSV file.

```python
page2= requests.get("https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating")
page2
```

```python
s2=BeautifulSoup(page2.content)
```

```python
movie_name=[]
rating=[]
year=[]
```

```python
movie_data = s2.findAll('div', attrs= {'class': 'lister-item mode-advanced'})
```

```python
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
    
    rate = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    rating.append(rate)
    
    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
    year.append(year_of_release)
    
```

```python
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of release': year, 'Movie Rating': rating})
```

```python
pd.set_option('display.max_rows', None)
```

### converting the dataframe to csv file

```python
movie_DF.to_csv('MovieRatings.csv')
```

```python
df=pd.read_csv('MovieRatings.csv')
df
```

The above csv file shows the top 100 movies according to IMDb ratings.

# 3. Python program to display IMDB’s Top rated 100 Indian movies’ data (i.e.Name, IMDB rating, Year of release) and saving it in form of a CSV file.

```python
page3= requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page3
```

```python
s3=BeautifulSoup(page3.content)
```

```python
Movie_name=[]
Rating=[]
Year=[]
```

```python
movie_Data = s3.find('tbody', {"class": 'lister-list' }).findAll("tr")
```

```python
for store in movie_Data:
    name1 = store.find('td', class_="titleColumn").find("a").get_text()
    Movie_name.append(name1)
    
    rate1 = store.find('td', class_ = 'ratingColumn imdbRating').find("strong").get_text()
    Rating.append(rate1)
    
    year_of_release1 = store.find('span', class_ = 'secondaryInfo').text.replace('(', '').replace(')', '')
    Year.append(year_of_release1)
```

```python
movie_df1 = pd.DataFrame({'Name of movie': Movie_name, 'Year of release': Year, 'Movie Rating': Rating})
```

```python
pd.set_option('display.max_rows', None)
```

### Converting dataframe to csv file

```python
movie_df1.to_csv('IndianMovieRatings.csv')
```

```python
DF=pd.read_csv('IndianMovieRatings.csv')
DF
```

The above csv file shows the top 100 Indian movies according to IMDb ratings.

# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
page4= requests.get("https://bookpage.com/reviews")
page4
```

```python
s4=BeautifulSoup(page4.content)
```

```python
book_name=[]
author_name=[]
genre=[]
review=[]
```

```python
book_data = s4.findAll('div', {"class": "flex-article-content" })
```

```python
for store in book_data:
    book = store.find('h4', class_="italic").a.text.replace('★', '')
    book_name.append(book)
    
    author = store.find('p', class_ = 'sans bold').text.replace('\n', '')
    author_name.append(author)
    
    genre_name = store.find('p', class_ = 'genre-links hidden-phone').a.text.replace('\n', '')
    genre.append(genre_name)
    
    book_review=store.find('p', class_ = 'excerpt').text.replace('\n', '')
    review.append(book_review)
```

```python
book_df = pd.DataFrame({'Book Name': book_name, 'Author': author_name, 'Genre': genre, 'Review': review})
```

```python
pd.set_option('display.max_rows', None)
pd.options.display.max_colwidth = 500
```

```python
book_df = book_df.drop([0,2], axis=0)
```

```python
new_book_df= book_df.reset_index(drop=True)
```

```python
new_book_df1=new_book_df[2:7]
```

```python
new_book_df1= new_book_df[2:7].reset_index(drop=True)
new_book_df1
```

The above 'new_book_df1' dataframe shows the details of five books.

# 5. Python program to scrape cricket rankings from ‘www.icc-cricket.com’

### <br>i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
pg_5= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
pg_5
```

```python
s5=BeautifulSoup(pg_5.content)
```

```python
table=s5.find('table', {'class': 'table'})
```

```python
headers=[]
```

```python
for i in table.find_all('th'):
    title=i.text.strip().replace("\n", " ")
    headers.append(title)
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(columns=headers)
for row in table.find_all('tr')[1:]:
    data=row.find_all('td')
    row_data=[td.text.replace("\n", " ") for td in data]
    length=len(df1)
    df1.loc[length]=row_data
```

```python
df1.head(10)
```

The above is the dataset showing top 10 ODI teams(Men's).

# 7. Python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data includes Product Name, Price, Image URL and Average Rating.

```python
page7= requests.get("https://www.amazon.in/Smartphones-Under-%E2%82%B920-000-Basic-Mobiles/s?rh=n%3A1805560031%2Cp_36%3A-2000000")
page7
```

```python
s7=BeautifulSoup(page7.content)
```

```python
product_name=[]
item_price=[]
img_url=[]
avg_rating=[]
```

```python
phone_data = s7.findAll('div', {"class": "a-section a-spacing-medium" })
```

```python
for store in phone_data:
    name = store.find('span', class_="a-size-base-plus a-color-base a-text-normal").get_text()
    product_name.append(name)
    
    price = store.find('span', class_ = 'a-price-whole').get_text()
    item_price.append(price)
```

```python
img_url=[]
img=s7.find_all('img', class_='s-image')
for i in img:
    img_url.append(i.get('src'))
```

```python
for x in s7.find_all('span', class_= 'a-icon-alt'):
    avg_rating.append(x.text)
```

```python
new_avg_rating=avg_rating[0:24]
```

```python
print(len(item_price), len(img_url), len(new_avg_rating), len(product_name)  )
```

```python
phone_df = pd.DataFrame({'Product Name': product_name, 'Price(in Rs.)': item_price, 'Phone Image URL': img_url, 'Rating':new_avg_rating })
```

```python
phone_df
```

The above dataframe 'phone_df' shows the details of the phoness below Rs. 20,000 on Amazon.in

# 8. Python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, SanFrancisco. Extracting data about 7 day extended forecast display for the city. This data includes period, short description, temperature and description.


```python
page8= requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJbbTbUzbIU")
page8
```

```python
s8=BeautifulSoup(page8.content)
```

```python
period=[]
short_description=[]
temp_hl=[]
```

```python
week = s8.find('div', {'id' : "seven-day-forecast-container"} )
```

```python
for item in week:
    day_time =item.find_all('p', class_='period-name')
    for i in day_time:
        period.append(i.text.replace('\n', " "))
period
```

```python
for item in week:
    short_desc =item.find_all('p', class_='short-desc')
    for i in short_desc:
        short_description.append(i.text.replace('\n', " "))
short_description
```

```python
for item in week:
    temp =item.find_all('p', class_='temp')
    for i in temp:
        temp_hl.append(i.text.replace('\n', " "))
temp_hl
```

```python
forecast=s8.find('div', class_="panel-body")
```

```python
forecast_details=[] 
detailed_weather= s8.find_all(class_ = 'col-sm-10')
for k in detailed_weather:
    forecast_details.append(k.text.replace('\n', " "))  
```

```python
forecast_details
```

```python
weather_df = pd.DataFrame({'Period': period, 'Short Description': short_description, 'Temperature': temp_hl, 'Detailed Forecast':forecast_details[0:9] })
```

```python
weather_df
```

The above dataframe 'weather_df' shows the weather forecast with period, short description, temperature and detailed Forecast.
